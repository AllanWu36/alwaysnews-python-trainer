## 学习一个程序之前，我们先需要理解什么是程序

在学习python之前，我们先学习更加宽泛的概念———什么是程序，python也只是程序的一种。

狭义的说，程序是一种利用机器完成的一系列序列化的操作。一个典型的特征就是：由两个部分组成： 数据和逻辑。

所以我们学习程序，主要就是学习如何管理数据和管理逻辑。

数据，无外乎就是这么几个东西：数据类型、数据结构、数据操作等。

逻辑，无外乎也就是这么几个东西：流程控制、函数、还有一些高级的操作。

## 程序的逻辑和数据在机器中如何运行

更深入一些，我们要了解数据和逻辑在计算机里是如何工作的

我们大概知道一个计算机大概有两个存储的设备，一个叫内存，一个是磁盘（手机上叫存储）。

这两个设备的关系可以理解为一个是菜板（指内存），一个是冰箱（指磁盘）。

当你真的运行一个程序，就像做菜一样，要把材料（数据和逻辑）从冰箱（磁盘）里拿到菜板（内存）上处理。  

冰箱（磁盘）的大小决定了你能装多少食材（安装多少程序）。

但真的做菜的时候，需要把食材都拿到菜板上进行处理，而菜板（内存）的大小就决定了你能同时做几个菜，也就是做菜的效率（运行程序的效率）。
